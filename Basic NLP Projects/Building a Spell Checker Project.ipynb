# 1)Finding Probability Distribution in a Text File

```python
import re
from tqdm import tqdm
```

```python
with open('big.txt', 'r') as fd:
    lines = fd.readlines()
    words = []
    for line in lines:
        words += re.findall('\w+' , line.lower())
len(words)
```

```python
print(words[:100])
```

```python
vocab = list(set(words))
print(len(vocab))
```

```python
words.count('the')#shows how many times 'the' appears in the whole text or corpus
```

```python
words.count('sherlock')
```

```python
len(words)
```

```python
#Probaility of 'The'
words.count('the') / len(words) 
```

```python
#we need to do it for all the words
word_probability = {}
for word in tqdm(vocab):
    word_probability[word] = float(words.count(word)/len(words))
    
```

# 2)Spell Checking Architecture

- We can Perform Four Basic operations on the text which are  add,replace,remove,swap.

## Splitting

```python
def split(word):    
    parts = []
    for i in range(len(word)+1):
        parts += [(word[: i] , word[i :])]
    return parts
split('loave')
```

#### Delete
### 'loave' -> 'love'

```python
for left,right in split('loave'):
    print(left +'||' +  right)
```

```python

def delete(word):
    output = []
    for left,right in split(word):
        output.append(left + right[1:] )
    return output
delete('loave')
```

### Swap
### 'lvoe' -> 'love'

```python
def swap(word):
    output = []
    for left,right in split(word):
        if (len(right) > 1):
            output.append(left + right[1] + right[0]  + right[2:])
    return output
swap('lvoe')
```

### Replace
### 'lave' -> 'love'

```python
def replace(word):
    characters = 'abcdefghijklmnopqrstuvwxyz'
    output = []
    for left, right in split(word):
        for char in characters:
            output.append(left + char + right[1:])
    return output
replace('lave')
#len(replace('lave'))
```

### Insert
### 'lve' -> 'love'

```python
def insert(word):
    characters = 'abcdefghijklmnopqrstuvwxyz'
    output = []
    for left,right in split(word):
        for char in characters:
            output.append(left + char + right)
    return output
insert('lve')
```

# Predicting the Correct Word(Level 1)

```python
word = 'loave'
print(len(insert(word)), len(delete(word)), len(swap(word)), len(replace(word)))
```

```python
def edit(word):
    return set(insert(word)+delete(word)+swap(word)+replace(word))# all the possible combination words
```

```python
suggested_word = edit('loave')
```

```python
output = []
for wrd in suggested_word:
    if wrd in word_probability.keys():#we are taking word_proability to check whether we have correct word and .keys is because it is dictionary
        output.append([wrd,word_probability[wrd]])#check which word has which has more probability
```

- Prediting the word

```python
def spell_check_edit_1(word, count = 5):
    output = []
    suggested_word = edit(word)
    for wrd in suggested_word:
        if wrd in word_probability.keys():
            output.append([wrd,word_probability[wrd]])
    return list(pd.DataFrame(output , columns = ['word' , 'prob']).sort_values(by = 'prob', ascending = False).head(count)['word'].values)
```

# Finding the Prediction (Level - 2)

```python
def spell_check_edit_2(word, count = 5):
    
    output = []
    suggested_words = edit(word)       # Level one Edit
    
    for e1 in edit(word):
        suggested_words = suggested_words.union(edit(e1))    # Second Level Edit 
    
    suggested_words = list(set(suggested_words))
    
    for wrd in suggested_words:
        if wrd in word_probability.keys():
            output.append([wrd, word_probability[wrd]])
    return list(pd.DataFrame(output, columns = ['word','prob']).sort_values(by = 'prob', ascending = False).head(count)['word'].values)
        
spell_check_edit_2('fameli')
```
